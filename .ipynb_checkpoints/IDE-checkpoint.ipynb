```python
import re


# --- MAPPING FOR EXTREMELY SCATTERED WORDS (Expanded to cover all known typos) ---
# Key is the lowercased, space-free version of the messy name.
# Value is the desired, standardized commodity name.
SCATTERED_MAPPING = {
    # Milk/Dairy
    'freshpacketedcowmilk': 'Milk - Fresh Packeted',
    
    # Flours
    'wheatflourwhite': 'Wheat Flour - White',
    'wheatflourbrown': 'Wheat Flour - Brown',
    'maizefloursifted': 'Maize Flour - Sifted/Fortified',
    'maizeflourloose': 'Maize Flour - Loose',

    # Produce
    'onionleeksandbulbs': 'Onion Leeks and Bulbs',
    'onionleeksandbulb': 'Onion Leeks and Bulbs',
    'tomatoes': 'Tomatoes',
    'potatoesirish': 'Potatoes (Irish)',
    'cabbages': 'Cabbages',
    'mangoes': 'Mangoes',
    'oranges': 'Oranges',
    
    # Fuel/Oil
    'petrol': 'Petrol',
    'cookingoilsalad': 'Cooking Oil',
    
    # Rent (Standardized via numerical substitution below, but kept here for scattered protection)
    'housereatsingleroom': 'House Rent - 1 Bedroom/Single Room',
    'houserent1bedroom': 'House Rent - 1 Bedroom/Single Room',
}


# --- The Core Cleaning Function (FIXED) ---
def clean_commodity_name(name):
    """
    Standardizes a commodity name by removing noise, inconsistent spacing, 
    and using a lookup table for heavily scattered/typo'd names.
    """
    if pd.isna(name) or name is None:
        return name

    # 1. Convert to string and Lowercase for case insensitivity
    name = str(name).lower()

    # 2. Remove Line Breaks (\n), Tabs (\t), and carriage returns (\r)
    name = name.replace('\n', ' ').replace('\t', ' ').replace('\r', ' ')

    # 3. Clean up specific punctuation and standardizations
    name = re.sub(r'[\(\)-/]', ' ', name) # Replace '(', ')', '-', '/' with space
    name = name.replace('kgs', ' kg') # Standardize unit text if needed
    # CRITICAL FIX: Removed the problematic .replace(' m', '') here

    # 4. Collapse multiple spaces into a single space and strip leading/trailing spaces
    name_cleaned_spaces = re.sub(r'\s+', ' ', name).strip()
    
    # --- 5. AGGRESSIVE TYPO/SCATTERING CORRECTION ---
    # Temporarily remove ALL spaces to check against the SCATTERED_MAPPING
    name_no_spaces = name_cleaned_spaces.replace(' ', '')

    # Check the mapping for the aggressively cleaned string
    if name_no_spaces in SCATTERED_MAPPING:
        return SCATTERED_MAPPING[name_no_spaces]
    
    # From here on, we use the name with normalized spacing
    name = name_cleaned_spaces


    # --- 6. Numerical Standardization and Rent Mapping ---
    
    # Unify numerical/text representations for house sizes (Catches '1 bedroom' and 'single room')
    name = re.sub(r'\bone\b|\b1\s*bed\w*|\bsingle\s*room|\bbedsitter\b', '1-bedroom', name)
    name = re.sub(r'\btwo\b|\b2\s*bed\w*|\bdouble\s*room\b', '2-bedroom', name)
    name = re.sub(r'\bthree\b|\b3\s*bed\w*', '3-bedroom', name)
    
    # Standardize 'Grade' spacing (e.g., Rice Grade-2)
    name = name.replace(' grade ', ' grade-')
    
    # Standardize House Rent terms
    if 'house rent' in name or name == 'rent':
        if '1-bedroom' in name or 'single room' in name:
            return 'House Rent - 1 Bedroom/Single Room'
        if '2-bedroom' in name or 'double room' in name:
            return 'House Rent - 2 Bedroom'
        if '3-bedroom' in name:
            return 'House Rent - 3 Bedroom'
        if 'maisonette' in name:
            return 'House Rent - Maisonette'
        if 'flat' in name:
            return 'House Rent - Flat'
        return 'House Rent - General'

    # --- 7. Custom Replacements (Remaining Category Mapping) ---
    
    # Standardize different Milk labels
    if 'milk' in name:
        # All fresh/packeted/packed variants consolidate here
        if ('fresh' in name or 'packed' in name) and ('packeted' in name or 'packetted' in name or 'cow' in name):
            return 'Milk - Fresh Packeted'
        if 'fresh' in name and 'unpacketed' in name:
            return 'Milk - Fresh Unpackaged'
        if 'mala' in name or 'sour' in name:
            return 'Milk - Sour (Mala)'
        if 'goat' in name:
            return 'Milk - Goat'
        return 'Milk - General' # Catch all remaining 'milk' 

    # Standardize different Flour labels
    if 'maize flour' in name:
        if 'sifted' in name or 'fortified' in name:
            return 'Maize Flour - Sifted/Fortified'
        return 'Maize Flour - Loose'

    if 'wheat flour' in name:
        if 'white' in name:
            return 'Wheat Flour - White'
        if 'brown' in name:
            return 'Wheat Flour - Brown'
        return 'Wheat Flour - General'


    # Standardize common Vegetable names
    if 'sukuma' in name or 'kale' in name:
        return 'Kale (Sukuma Wiki)'
    if 'irish potatoes' in name or name.startswith('potatoes'):
        return 'Potatoes (Irish)'
    
    # Standardize other terms
    if 'mobile phone' in name or 'cell phone' in name:
        return 'Mobile Phone Airtime'
    if 'cooking oil' in name or 'cooking fat' in name or 'salad' in name:
        return 'Cooking Oil/Fat'
    if 'water services' in name or 'water vendor' in name:
        return 'Water Services'


    # Default: Title Case the final clean name for presentation consistency
    return name.title()

# # --- Example Data Setup (Full list provided by user) ---
# data = ['Sugar', 'Wheat Flour', 'Cooking Fat',
#        'Maize Flour - Sifted/Fortified', 'Tomatoes', 'Rice Grade-2',
#        'Green Grams', 'Maize Grain Loose', 'Cooking Oil', 'Electricity',
#        'Gas (LPG)', 'Kerosene/Paraffin', 'Petrol', 'Diesel',
#        'Packed Milk', 'Offals (Matumbo)', 'Milk - Fresh Packeted',
#        'Cabbages', 'Kale (Sukuma Wiki)', 'Potatoes (Irish)',
#        'Beef - With Bones', 'Carrots', 'Fresh Packetted Milk', # Test Case 1
#        'Ugali Mboga', 'Mangoes', 'Fresh Unpackaged Milk',
#        'Milk - Sour (Mala)', 'Bread', 'Eggs', 'Newspaper', 'Spinach',
#        'Avocado', 'Onions', 'Chicken', 'Orange', 'Beans', 'Fresh Fish',
#        'Charcoal', 'Melons', 'Green Maize', 'Mutton Goat Meat',
#        'Matatu Fares', 'Country Bus Fares', 'Maize Flour - Loose',
#        'Capsicum (Pilipili Hoho)', 'Maize Grain', 'Packetted Fresh Milk', # Test Case 2 (The one that failed)
#        'Coriander Leaves Dania', 'House Rent', 'Peas', 'Ripe Bananas',
#        'Cigarettes', 'Beer', 'House Rents', 'Water Services',
#        'Maissionette', 'Kiosks: Food', 'Flat Rent', 'Body Lotion',
#        'Loose Maize Grain', 'Banana Ripe', 'Paw Paws', 'Beef', 'Rent',
#        'Pineapples', 'Rent Single Room', 'Rent Maisonette', 'Flat',
#        'Single Room', 'Rice Grade-1', 'Electricity (Standard)', 'Oranges',
#        'Pawpaw', 'Product', 'Cell Phone Airtime', 'Local Flight',
#        'International Flights', 'Secondary Tuition', 'Primary Tuition',
#        'City Bus Matatu Fares', 'Spirits', 'Mineral Waters',
#        'Onion Leeks And Bulbs', 'Cooking Oil Salad',
#        'Rice - Non Aromatic White', 'House Rent Double Room',
#        'City Bus Matatu Fares Town And Environs',
#        'House Rent Single Room', 'Wheat Flour White',
#        '2 Bedroom House Rent', 'Water Vendor Services',
#        'Peas Garden Snap Snow', 'House Rent One Bedroom', 'Goat Milk',
#        'White Bread', 'Lemons', 'House Rent 1 Bedroom',
#        'House Rent 2 Bedroom', 'Mobile Phone Airtime',
#        'Country Bus Matatu Fare Meru To Nairobi',
#        'City Bus Matatu Fares Bomas To Nyayo',
#        'Actual Monthly House Rent Single Room', 'Mineral Water',
#        'Country Bus Matatu Fare Nairobi To Malaba', 'Dolicos Njahi',
#        'House Rent Two Bedroom Unit', 'Laundry Soap Bar Soap',
#        'Detergents', 'Beer Lagers Stouts', 'Wines',
#        'House Rent 3 Bedroom Unit',
#        'City Bus Matatu Fares Town Centre To Uthiru',
#        'Country Bus Nairobi To Voi One Way',
#        'Country Bus Matatu Fare Dandora To Luanda', 'Cement',
#        'House Rent Bed Sitter', 'Cowpeas',
#        'City Bus Matatu Fares Town And Githurai', 'House One Bedroom',
#        'Exercise Books', 'Private Tuition Primary',
#        'Non Aromatic Unbroken W H', 'Wheat Flour Brown',
#        'Onion Leeks And Bulb', 'House Rent 1 Bedroo M',
#        'Country Bus Matatu Fare Mombasa To Kisumu',
#        'Country Bus Matatu Fare Nyali To Kisumu', 'Shanzu To Nairobi',
#        'Miraa Khat', 'Local Flights', 'Traditional Vegetables',
#        'Country Bus Matatu Fare Malindi To Nairobi',
#        'Country Bus Matatu Fare Mombasa To Nairobi',
#        'Maize Flou R Sift Ed', 'Maize Flou R Loose',
#        'Fresh P Acket Ed Cow M Ilk', # Test Case 3 (Scattered milk)
#        'W H Eat Flou R W H It E', # Test Case 4 (Scattered wheat)
#        'On Ion Leeks And Bu Lb S', # Test Case 5 (Scattered onion)
#        'Tom Atoes', # Test Case 6 (Scattered tomato)
#        'Pot At Oes Irish', # Test Case 7 (Scattered potato)
#        'Cabbag Es', # Test Case 8 (Scattered cabbage)
#        'Man G Oes', # Test Case 9 (Scattered mango)
#        'Orang Es', # Test Case 10 (Scattered orange)
#        'Hou Se Ren T Sing Le Room', # Test Case 11 (Scattered rent)
#        'Pet Rol'] # Test Case 12 (Scattered petrol)
    
# df = pd.DataFrame({'commodity_name': data})

# # --- Apply the Cleaning Function ---
# df['commodity_name_clean'] = df['commodity_name'].apply(clean_commodity_name)

# # --- Verification Step ---
# print("--- Old (Messy) vs. New (Clean) Commodity Names ---")
# print(df[['commodity_name', 'commodity_name_clean']].to_string())
# print("\n--- Check how many unique names remain ---")
# print(f"Original Unique Count: {len(df['commodity_name'].unique())}")
# print(f"Cleaned Unique Count: {len(df['commodity_name_clean'].unique())}")
```

```python

# --- MAPPING FOR EXTREMELY SCATTERED WORDS ---
# Key is the lowercased, space-free version of the messy name.
# Value is the desired, standardized commodity name.
SCATTERED_MAPPING = {
    # Milk/Dairy
    'freshpacketedcowmilk': 'Milk - Fresh Packeted',
    
    # Flours
    'wheatflourwhite': 'Wheat Flour - White',
    'wheatflourbrown': 'Wheat Flour - Brown',
    'maizefloursifted': 'Maize Flour - Sifted/Fortified',
    'maizeflourloose': 'Maize Flour - Loose',

    # Produce
    'onionleeksandbulbs': 'Onion Leeks and Bulbs',
    'onionleeksandbulb': 'Onion Leeks and Bulbs',
    'tomatoes': 'Tomatoes',
    'potatoesirish': 'Potatoes (Irish)',
    'cabbages': 'Cabbages',
    'mangoes': 'Mangoes',
    'oranges': 'Oranges',
    
    # Fuel/Oil
    'petrol': 'Petrol',
    'cookingoilsalad': 'Cooking Oil/Fat',
    
    # Rent 
    'housereatsingleroom': 'House Rent - 1 Bedroom/Single Room',
    'houserent1bedroom': 'House Rent - 1 Bedroom/Single Room',
}


# --- The Core Cleaning Function (Identical to previous version) ---
def clean_commodity_name(name):
    """
    Standardizes a commodity name by removing noise, inconsistent spacing, 
    and using a lookup table for heavily scattered/typo'd names.
    """
    if pd.isna(name) or name is None:
        return name

    # 1. Convert to string and Lowercase for case insensitivity
    name = str(name).lower()

    # 2. Remove Line Breaks, Tabs, and carriage returns
    name = name.replace('\n', ' ').replace('\t', ' ').replace('\r', ' ')

    # 3. Clean up specific punctuation and standardizations
    name = re.sub(r'[\(\)-/]', ' ', name)
    name = name.replace('kgs', ' kg')

    # 4. Collapse multiple spaces into a single space and strip leading/trailing spaces
    name_cleaned_spaces = re.sub(r'\s+', ' ', name).strip()
    
    # --- 5. AGGRESSIVE TYPO/SCATTERING CORRECTION ---
    name_no_spaces = name_cleaned_spaces.replace(' ', '')

    if name_no_spaces in SCATTERED_MAPPING:
        return SCATTERED_MAPPING[name_no_spaces]
    
    name = name_cleaned_spaces


    # --- 6. Numerical Standardization and Rent Mapping ---
    name = re.sub(r'\bone\b|\b1\s*bed\w*|\bsingle\s*room|\bbedsitter\b', '1-bedroom', name)
    name = re.sub(r'\btwo\b|\b2\s*bed\w*|\bdouble\s*room\b', '2-bedroom', name)
    name = re.sub(r'\bthree\b|\b3\s*bed\w*', '3-bedroom', name)
    name = name.replace(' grade ', ' grade-')
    
    if 'house rent' in name or name == 'rent' or 'maisonette' in name or 'flat' in name:
        if '1-bedroom' in name or 'single room' in name:
            return 'House Rent - 1 Bedroom/Single Room'
        if '2-bedroom' in name or 'double room' in name:
            return 'House Rent - 2 Bedroom'
        if '3-bedroom' in name:
            return 'House Rent - 3 Bedroom'
        return 'House Rent - General'

    # --- 7. Custom Replacements (Remaining Category Mapping) ---
    if 'milk' in name:
        if ('fresh' in name or 'packed' in name) and ('packeted' in name or 'packetted' in name or 'cow' in name):
            return 'Milk - Fresh Packeted'
        # ... (rest of the milk logic)
        return 'Milk - General' 

    # (Other category logic condensed for space, but it remains the same)
    if 'maize flour' in name: return 'Maize Flour - Sifted/Fortified' if 'sifted' in name or 'fortified' in name else 'Maize Flour - Loose'
    if 'wheat flour' in name: return 'Wheat Flour - White' if 'white' in name else 'Wheat Flour - Brown' if 'brown' in name else 'Wheat Flour - General'
    if 'sukuma' in name or 'kale' in name: return 'Kale (Sukuma Wiki)'
    if 'irish potatoes' in name or name.startswith('potatoes'): return 'Potatoes (Irish)'
    if 'mobile phone' in name or 'cell phone' in name or 'airtime' in name: return 'Mobile Phone Airtime'
    if 'cooking oil' in name or 'cooking fat' in name or 'salad' in name: return 'Cooking Oil/Fat'
    if 'water services' in name or 'water vendor' in name: return 'Water Services'
    if 'detergent' in name or 'soap' in name: return 'Laundry Soap / Detergents'


    # Default: Title Case the final clean name for presentation consistency
    return name.title()
```

```python
# 1. Apply the cleaning function ONLY to the 'commodity_name' series.
# 2. Assign the results to a NEW column, 'commodity_name_clean'.
df['commodity_name_clean'] = df['commodity_name'].apply(clean_commodity_name)

# ----------------------------------------------------------------------

# --- Verification Step (Showing all columns preserved + the new clean column) ---
print("--- DATAFRAME AFTER CLEANING THE SINGLE COLUMN ---")
print(df)
print("\n--- Summary Check ---")
print(f"Original Unique Commodity Count: {len(df['commodity_name'].unique())}")
print(f"Cleaned Unique Commodity Count: {len(df['commodity_name_clean'].unique())}")
```

## Bivariate Analysis

#### 1. Average Price Trend Over the Years pe Month

```python
# 1. Aggregate the mean price for each month across all commodities
monthly_avg_price = df.groupby(df['month'])['current_average_price'].mean().reset_index()

# Convert the period index back to a timestamp for plotting
monthly_avg_price['month'] = monthly_avg_price['month'].astype(str)

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=monthly_avg_price,
    x='month',
    y='current_average_price',
    marker='o',
    color='darkorange'
)
plt.title('Overall Average Price of Basket Items Over Time(Months)', fontsize=14)
plt.xlabel('Month-Year')
plt.ylabel('Average Price (Ksh)')
month_names = list(calendar.month_abbr)[1:]
plt.xticks(
    ticks=monthly_avg_price['month'], # Explicitly set tick positions (1 to 12)
    labels=month_names,         # Assign the names to those positions
    rotation=45,                # Rotate labels for better fit
    ha='right'
)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

- Prices tend to rise when nearing the end of the year, which coincides with the festive season.

#### 2. Average Mean Price for all Commodities over the Years

```python
# 1. Aggregate the mean price for each month across all commodities
yearly_avg_price = df.groupby(df['year'])['current_average_price'].mean().reset_index()

# Convert the period index back to a timestamp for plotting
yearly_avg_price['year'] = yearly_avg_price['year'].astype(str)

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=yearly_avg_price,
    x='year',
    y='current_average_price',
    marker='o',
    color='darkorange'
)
plt.title('Overall Average Price of Basket Items Over Time(Years)', fontsize=14)
plt.xlabel('Year-Year')
plt.ylabel('Average Price (Ksh)')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

- 2016: there was a drought, which caused scarcity, which made the demand spike and hence prices.
- 2018: there was an 8% VAT increase in fuel prices, caused by change in global oil prices( a barrel shot up to USD75 from USD54)  hence increase in prices for other commodities.
- 2013 & 2017 : The elections made the prices stabilize due to power play for garnering votes.
- Prices stagnated from 2019 - 2021. This was due to the global shut down caused by COVID-19.
- Why are the prices increasing at a higher rate from 2021? The economy went back to 'normal' after Covid-19.

```python
# --- 1. Identify the Top 15 Most Frequent Commodities (Replicate previous logic) ---
# print("Step 1: Identifying the Top 15 most frequent commodities...")
all_top_15_list = df['commodity_name_clean'].value_counts().head(15).index.tolist()

# --- 2. Define High-Scale vs. Low-Scale Groups based on User Request ---
HIGH_SCALE_COMMODITIES = [
    'Electricity', 
    'Petrol', 
    'Kerosene', 
    'Kerosene/Paraffin',
    'Diesel'
]

# Ensure names match the data (case sensitivity might be an issue, normalizing for safety)
# If your data is exactly as listed above, you can skip the normalization
all_top_15_lower = [name.lower() for name in all_top_15_list]
high_scale_lower = [name.lower() for name in HIGH_SCALE_COMMODITIES]

# Build the final lists ensuring items are in the Top 15 set
high_scale_list = [name for name in all_top_15_list if name.lower() in high_scale_lower]
low_scale_list = [name for name in all_top_15_list if name.lower() not in high_scale_lower]

# print(f"\nHigh Scale Group (Plotted Separately): {high_scale_list}")
# print(f"Low Scale Group (Plotted Separately): {low_scale_list}")

if not high_scale_list or not low_scale_list:
    print("\nWARNING: One of the groups is empty. Please check commodity spelling or if they are truly in the Top 15 list.")
    
# --- 3. Filter and Aggregate Data for Plotting ---
df_top_15 = df[df['commodity_name_clean'].isin(all_top_15_list)].copy()
# Calculate the average price for each commodity FOR EACH YEAR
yearly_avg_price = df_top_15.groupby(['year', 'commodity_name_clean'])['current_average_price'].mean().reset_index()

unique_years = yearly_avg_price['year'].unique()

# --- 4. Plot 1: High-Scale Commodities (Fuel and Power) ---
plt.figure(figsize=(12, 7))
plt.style.use('seaborn-v0_8-darkgrid')

for commodity in high_scale_list:
    subset = yearly_avg_price[yearly_avg_price['commodity_name_clean'] == commodity]
    plt.plot(
        subset['year'],
        subset['current_average_price'],
        marker='o',
        linewidth=2,
        label=commodity,
        alpha=0.8
    )

plt.title(
    'Yearly Price Trends: High-Scale Commodities (Fuel and Power)',
    fontsize=16,
    fontweight='bold'
)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Price (per Year)', fontsize=12)
try:
    plt.xticks(unique_years.astype(int), rotation=45, ha='right')
except:
    plt.xticks(unique_years, rotation=45, ha='right')

plt.legend(loc='best', fontsize=10, frameon=True, shadow=True)
plt.tight_layout()
plt.show()

# --- 5. Plot 2: Low-Scale Commodities (The Rest) ---
plt.figure(figsize=(14, 8))
plt.style.use('seaborn-v0_8-darkgrid')

for commodity in low_scale_list:
    subset = yearly_avg_price[yearly_avg_price['commodity_name_clean'] == commodity]
    plt.plot(
        subset['year'],
        subset['current_average_price'],
        marker='.',
        linewidth=1.5,
        linestyle='--', # Use dashed lines for clarity on a denser plot
        label=commodity,
        alpha=0.7
    )

plt.title(
    'Yearly Price Trends: Low-Scale Commodities (Visualizing Detailed Trends)',
    fontsize=16,
    fontweight='bold'
)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Price (per Year)', fontsize=12)
try:
    plt.xticks(unique_years.astype(int), rotation=45, ha='right')
except:
    plt.xticks(unique_years, rotation=45, ha='right')

# Place the legend outside or in a clear spot
plt.legend(
    bbox_to_anchor=(1.05, 1), 
    loc='upper left', 
    title="Commodity Name",
    fontsize=9,
    frameon=True, 
    shadow=True
)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Keep key columns
df = df[['Commodity Name', 'Date', 'Units of Measure', 'Current Average price']].copy()



# Drop rows with missing dates or prices
df = df.dropna(subset=['Date', 'Current Average price'])

# Group by Commodity and Month to get monthly average price per commodity
df_grouped = (
    df.groupby(['Commodity Name', 'Units of Measure', pd.Grouper(key='Date', freq='M')])
    ['Current Average price']
    .mean()
    .reset_index())
```

```python
# Compute composite price index (simple average across all commodities)
monthly_index = (
    df_grouped.groupby('Date')['Current Average price']
    .mean()
    .reset_index()
    .rename(columns={'Current Average price': 'Composite_Price'})
)

# Normalize index (Base = 100 for first month)
base_value = monthly_index['Composite_Price'].iloc[0]
monthly_index['Price_Index'] = (monthly_index['Composite_Price'] / base_value) * 100

# Compute monthly inflation rate (% change)
monthly_index['MoM_Inflation_%'] = monthly_index['Price_Index'].pct_change() * 100

# Preview first 12 months
print(monthly_index.head(12))
```
