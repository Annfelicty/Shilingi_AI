```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
import calendar
```

```python
df = pd.read_excel("data1.xlsx")
df.head(20)
```

```python
# 20 most common commodities
df_valuecounts = df['Commodity Name'].value_counts().head(20)
df_valuecounts 
```

```python
# checking the info
df.info()

# we will need to convert some col into integers, and the date to date-time
```

```python
 # remove leading/trailing spaces
df.columns = df.columns.str.strip() 
# replace spaces with underscores
df.columns = df.columns.str.replace(' ', '_')  

df.columns.tolist()
```

```python
# List of columns to convert
cols_to_convert = ['Average_Price_Previous_Year','Average_Price_Previous_Month','Current_Average_price',
                   '%_monthly_change','%_yearly_change']

# Cleaning and converting each column
for col in cols_to_convert:
    df[col] = (df[col]
               .astype(str)                     # ensure string type
               .str.replace(',', '', regex=False)  # remove commas
               .str.replace('%', '', regex=False)  # remove percent signs
               .str.strip())                        # remove spaces
    df[col] = pd.to_numeric(df[col], errors='coerce')  # convert to numeric safely

# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# convert the columns to lowercase
df.columns = df.columns.str.lower()
df.columns.tolist()
```

```python
df.info()
```

```python
# very little in the grandscheme
df.duplicated().sum()
```

```python
# checking null values --- None is missing. Cleaning done
df.isna().sum()
```

```python
# checking the data column
df[df['date'].isna()]
```

```python
# it appears to have missing data across all columnns and we will drop this row
df.drop(1048, inplace=True)
```

```python
df[df['average_price_previous_month'].isna()]
```

```python
# Recompute missing % changes
df['%_monthly_change'] = df['%_monthly_change'].fillna(
    ((df['current_average_price'] - df['average_price_previous_month']) / df['average_price_previous_month']) * 100)

df['%_yearly_change'] = df['%_yearly_change'].fillna(
    ((df['current_average_price'] - df['average_price_previous_year']) / df['average_price_previous_year']) * 100)

# Recompute missing prices (inverse calculation)
df['average_price_previous_month'] = df['average_price_previous_month'].fillna(
    df['current_average_price'] / (1 + (df['%_monthly_change'] / 100)))

df['average_price_previous_year'] = df['average_price_previous_year'].fillna(
    df['current_average_price'] / (1 + (df['%_yearly_change'] / 100)))

# Check remaining nulls
df.isna().sum()
```

```python
df[df['%_yearly_change'].isna()]
```

```python
# dropping the null values
df = df.dropna()
df 
```

```python
# validating there are no missing values
df.isna().sum()
```

```python
df.columns
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

#### 3. KDE Plot for prices

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(
    data=df,
    x='current_average_price',
    fill=True,
    color='teal',
    alpha=0.6
)
plt.title('Distribution of Current Average Price Across All Data', fontsize=14)
plt.xlabel('Current Average Price')
plt.ylabel('Density')
plt.show()
```

```python
# --- 1. Group the data by commodity and calculate the standard deviation (volatility)
# We use the yearly change as a measure of long-term price fluctuation
volatility_df = df.groupby('commodity_name')['%_yearly_change'].std().reset_index()
volatility_df.columns = ['commodity_name', 'yearly_price_volatility']

# --- 2. Find the 10 commodities with the highest volatility (nlargest)
top_10_volatile = volatility_df.nlargest(10, 'yearly_price_volatility')

# --- 3. Find the 10 commodities with the lowest volatility (nsmallest)
# These are the commodities with the most stable prices.
bottom_10_stable = volatility_df.nsmallest(10, 'yearly_price_volatility')


print("--- Top 10 Most Volatile Commodities (Highest STD of Yearly Change) ---\n")
print(top_10_volatile)
print("\n" + "="*70 + "\n")
print("--- Top 10 Most Price-Stable Commodities (Lowest STD of Yearly Change) ---\n")
print(bottom_10_stable)
```

## Bivariate Analysis

#### 1. Average Price Trend Over Time

```python
# 1. Aggregate the mean price for each month across all commodities
monthly_avg_price = df.groupby(df['month'])['current_average_price'].mean().reset_index()

# Convert the period index back to a timestamp for plotting
monthly_avg_price['month'] = monthly_avg_price['month'].astype(str)

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=monthly_avg_price,
    x='month',
    y='current_average_price',
    marker='o',
    color='darkorange'
)
plt.title('Overall Average Price of Basket Items Over Time(Months)', fontsize=14)
plt.xlabel('Month-Year')
plt.ylabel('Average Price (Ksh)')
month_names = list(calendar.month_abbr)[1:]
plt.xticks(
    ticks=monthly_avg_price['month'], # Explicitly set tick positions (1 to 12)
    labels=month_names,         # Assign the names to those positions
    rotation=45,                # Rotate labels for better fit
    ha='right'
)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 1. Aggregate the mean price for each month across all commodities
yearly_avg_price = df.groupby(df['year'])['current_average_price'].mean().reset_index()

# Convert the period index back to a timestamp for plotting
yearly_avg_price['year'] = yearly_avg_price['year'].astype(str)

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=yearly_avg_price,
    x='year',
    y='current_average_price',
    marker='o',
    color='darkorange'
)
plt.title('Overall Average Price of Basket Items Over Time(Years)', fontsize=14)
plt.xlabel('Year-Year')
plt.ylabel('Average Price (Ksh)')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Keep key columns
df = df[['Commodity Name', 'Date', 'Units of Measure', 'Current Average price']].copy()



# Drop rows with missing dates or prices
df = df.dropna(subset=['Date', 'Current Average price'])

# Group by Commodity and Month to get monthly average price per commodity
df_grouped = (
    df.groupby(['Commodity Name', 'Units of Measure', pd.Grouper(key='Date', freq='M')])
    ['Current Average price']
    .mean()
    .reset_index())
```

```python
# Compute composite price index (simple average across all commodities)
monthly_index = (
    df_grouped.groupby('Date')['Current Average price']
    .mean()
    .reset_index()
    .rename(columns={'Current Average price': 'Composite_Price'})
)

# Normalize index (Base = 100 for first month)
base_value = monthly_index['Composite_Price'].iloc[0]
monthly_index['Price_Index'] = (monthly_index['Composite_Price'] / base_value) * 100

# Compute monthly inflation rate (% change)
monthly_index['MoM_Inflation_%'] = monthly_index['Price_Index'].pct_change() * 100

# Preview first 12 months
print(monthly_index.head(12))
```
