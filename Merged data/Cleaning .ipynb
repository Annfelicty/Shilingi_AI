**üéØ Goal:**  
Prepare both datasets for **accurate numeric analysis** and **time-based inflation modeling** by ensuring all columns are properly typed and free of formatting noise.


```python
# Step 7: Create Robust Numeric Converter
print("üîÑ CONVERTING STRINGS TO NUMERIC TYPES")
print("="*50)

def clean_and_convert_numeric(value):
    """
    Convert messy string numbers to proper numeric values
    Handles: commas, percentage signs, mixed text/numeric, headers
    """
    if pd.isna(value):
        return np.nan
    
    # Convert to string and clean
    str_val = str(value).strip()
    
    # Skip if it's clearly a header/description (contains words like 'price', 'change', 'average')
    header_keywords = ['price', 'change', 'average', 'weight', 'month', 'year']
    if any(keyword in str_val.lower() for keyword in header_keywords) and not any(c.isdigit() for c in str_val):
        return np.nan
    
    # Remove percentage signs and convert to decimal
    if '%' in str_val:
        str_val = str_val.replace('%', '').strip()
        # If it's a percentage, convert to decimal (e.g., 25% ‚Üí 0.25)
        try:
            numeric_val = float(str_val) / 100.0
            return numeric_val
        except:
            return np.nan
    
    # Remove commas (thousands separators)
    str_val = str_val.replace(',', '')
    
    # Remove extra spaces within numbers (e.g., "1 275.05" ‚Üí "1275.05")
    str_val = re.sub(r'(\d)\s+(\d)', r'\1\2', str_val)
    
    # Try to convert to float
    try:
        return float(str_val)
    except:
        return np.nan

def convert_numeric_columns(df, df_name):
    print(f"\nüî¢ CONVERTING {df_name.upper()} NUMERIC COLUMNS:")
    
    # Define which columns should be numeric based on our analysis
    if df_name == "CPI":
        numeric_cols = ['weight_pct', 'yoy_change_pct', 'yoy_change_pct_alt', 
                       'price_march_2023', 'monthly_change_pct', 'annual_change_pct']
    else:  # Commodity
        numeric_cols = ['base_price_2010', 'current_price', 'price_change_pct',
                       'monthly_change_pct', 'annual_change_pct']
    
    for col in numeric_cols:
        if col in df.columns:
            # Create backup before conversion
            df[f'{col}_backup'] = df[col]
            
            # Convert the column
            original_non_null = df[col].notna().sum()
            df[col] = df[col].apply(clean_and_convert_numeric)
            new_non_null = df[col].notna().sum()
            
            print(f"   {col}: {original_non_null} ‚Üí {new_non_null} non-null values")
            
            # Show conversion summary
            if new_non_null > 0:
                print(f"     Range: {df[col].min():.3f} to {df[col].max():.3f}")
    
    return df

# Apply conversion to both datasets
cpi_clean = convert_numeric_columns(cpi_clean, "CPI")
commod_clean = convert_numeric_columns(commod_clean, "Commodity Prices")
```

```python
# Step 8: Fix Date Columns
print("\nüìÖ FIXING DATE COLUMNS")
print("="*50)

def fix_date_columns(df, df_name):
    print(f"\nüìÖ CONVERTING {df_name.upper()} DATE COLUMN:")
    
    if 'date' in df.columns:
        # Convert to proper datetime
        original_non_null = df['date'].notna().sum()
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
        new_non_null = df['date'].notna().sum()
        
        print(f"   Date conversion: {original_non_null} ‚Üí {new_non_null} valid dates")
        if new_non_null > 0:
            print(f"   Date range: {df['date'].min()} to {df['date'].max()}")
        
        # Create time-based features
        df['year'] = df['date'].dt.year
        df['month'] = df['date'].dt.month
        df['quarter'] = df['date'].dt.quarter
        df['year_month'] = df['date'].dt.to_period('M')
        
        print(f"   Created time features: year, month, quarter, year_month")
    
    return df

cpi_clean = fix_date_columns(cpi_clean, "CPI")
commod_clean = fix_date_columns(commod_clean, "Commodity Prices")
```

```python
# Step 9: Verify Our Conversions
print("\n" + "="*60)
print("‚úÖ CONVERSION VERIFICATION")
print("="*60)

def verify_conversions(df, df_name):
    print(f"\nüìä {df_name.upper()} - DATA TYPES AFTER CONVERSION:")
    print(df.dtypes)
    
    print(f"\nüìà {df_name.upper()} - NUMERIC COLUMNS SUMMARY:")
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 0:
        display(df[numeric_cols].describe())

verify_conversions(cpi_clean, "CPI")
verify_conversions(commod_clean, "Commodity Prices")
```

```python
# Step 10: Investigate Extreme Values and Outliers
print("\nüîç INVESTIGATING EXTREME VALUES")
print("="*50)

def investigate_extreme_values(df, df_name):
    print(f"\nüìä {df_name.upper()} - EXTREME VALUE ANALYSIS:")
    
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    
    for col in numeric_cols:
        if col not in ['year', 'month', 'quarter']:  # Skip time features
            print(f"\nüìà Column: {col}")
            
            # Calculate robust statistics (less sensitive to outliers)
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Count outliers
            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
            outlier_pct = (len(outliers) / len(df)) * 100
            
            print(f"   Normal range: [{lower_bound:.2f}, {upper_bound:.2f}]")
            print(f"   Outliers: {len(outliers)} ({outlier_pct:.1f}%)")
            print(f"   Extreme samples: {outliers.head(5).tolist()}")
            
            # Show what these outliers are associated with
            if len(outliers) > 0 and 'broad_item_group' in df.columns:
                outlier_groups = df.loc[outliers.index, 'broad_item_group'].value_counts().head(3)
                print(f"   Top groups with outliers: {outlier_groups.to_dict()}")

investigate_extreme_values(cpi_clean, "CPI")
investigate_extreme_values(commod_clean, "Commodity Prices")
```

```python
# Step 12: Smart Outlier Handling Strategy
print("\nüõ†Ô∏è IMPLEMENTING OUTLIER HANDLING STRATEGY")
print("="*50)

def handle_outliers_smart(df, df_name):
    print(f"\nüéØ {df_name.upper()} - OUTLIER HANDLING:")
    
    # Strategy: Cap extreme values at reasonable limits
    outlier_handling = {}
    
    if df_name == "CPI":
        outlier_handling = {
            'yoy_change_pct': (-50, 50),        # Cap at ¬±50% yearly change
            'yoy_change_pct_alt': (-50, 50),    # Same for alternative
            'monthly_change_pct': (-30, 30),    # Cap at ¬±30% monthly
            'annual_change_pct': (-50, 50),     # Cap at ¬±50% annual
            'price_march_2023': (0, 10000),     # No negative prices, cap at 10,000
            'weight_pct': (0, 100)              # Weights should be 0-100%
        }
    else:  # Commodity
        outlier_handling = {
            'base_price_2010': (0, 5000),       # No negative, cap at 5,000
            'current_price': (0, 5000),         # Same for current prices
            'price_change_pct': (-80, 500),     # Allow big increases but cap extremes
            'monthly_change_pct': (-50, 50),    # Cap monthly changes
            'annual_change_pct': (-50, 100)     # Allow 100% annual increases max
        }
    
    for col, (lower_bound, upper_bound) in outlier_handling.items():
        if col in df.columns:
            # Count values that will be capped
            below_cap = (df[col] < lower_bound).sum()
            above_cap = (df[col] > upper_bound).sum()
            
            # Apply capping
            df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
            
            print(f"   {col}: Capped {below_cap} low + {above_cap} high values")
            print(f"     New range: [{df[col].min():.2f}, {df[col].max():.2f}]")
    
    return df

# Apply outlier handling
cpi_clean = handle_outliers_smart(cpi_clean, "CPI")
commod_clean = handle_outliers_smart(commod_clean, "Commodity Prices")
```

```python
# Step 15: Final Missing Value Strategy
print("\nüîç FINAL MISSING VALUE ANALYSIS & STRATEGY")
print("="*50)

def analyze_and_handle_missing(df, df_name):
    print(f"\nüìä {df_name.upper()} - MISSING VALUE STRATEGY:")
    
    # Show current missingness
    missing = df.isna().sum()
    missing_pct = (df.isna().mean() * 100).round(2)
    missing_df = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
    missing_df = missing_df[missing_df['missing_count'] > 0].sort_values('missing_pct', ascending=False)
    
    if len(missing_df) > 0:
        print("Missing values before handling:")
        display(missing_df)
        
        # Strategy: For numeric columns with <30% missing, use median imputation
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        imputed_cols = []
        
        for col in numeric_cols:
            if col in missing_df.index and missing_df.loc[col, 'missing_pct'] < 30:
                before_count = df[col].isna().sum()
                median_val = df[col].median()
                df[col] = df[col].fillna(median_val)
                after_count = df[col].isna().sum()
                imputed_cols.append((col, before_count, median_val))
                print(f"   ‚úÖ {col}: Filled {before_count} missing values with median {median_val:.2f}")
        
        if not imputed_cols:
            print("   ‚úÖ No numeric columns with <30% missing needed imputation")
    
    return df

print("üîÑ Handling missing values for CPI...")
cpi_clean = analyze_and_handle_missing(cpi_clean, "CPI")

print("\nüîÑ Handling missing values for Commodity...")
commod_clean = analyze_and_handle_missing(commod_clean, "Commodity Prices")
```

```python
# Step 17: Final Cleanup - Remove Backup Columns
print("\nüßπ FINAL CLEANUP - REMOVING BACKUP COLUMNS")
print("="*50)

def remove_backup_columns(df, df_name):
    print(f"\nüìä {df_name.upper()} - REMOVING BACKUPS:")
    
    backup_cols = [col for col in df.columns if 'backup' in col]
    print(f"   Backup columns to remove: {backup_cols}")
    
    df_clean = df.drop(columns=backup_cols, errors='ignore')
    print(f"   Shape: {df.shape} ‚Üí {df_clean.shape}")
    print(f"   Remaining columns: {len(df_clean.columns)}")
    
    return df_clean

cpi_final = remove_backup_columns(cpi_clean, "CPI")
commod_final = remove_backup_columns(commod_clean, "Commodity Prices")
```

```python
# Step 18: Handle Remaining Date Issues
print("\nüìÖ HANDLING REMAINING DATE ISSUES")
print("="*50)

def handle_date_issues(df, df_name):
    print(f"\nüìä {df_name.upper()} - DATE CLEANUP:")
    
    # Count rows with missing dates
    missing_dates = df['date'].isna().sum()
    print(f"   Rows with missing dates: {missing_dates} ({missing_dates/len(df)*100:.1f}%)")
    
    # Option 1: Remove rows with missing dates (if small percentage)
    if missing_dates / len(df) < 0.15:  # Less than 15% missing
        df_clean = df.dropna(subset=['date'])
        print(f"   ‚úÖ Removed {missing_dates} rows with missing dates")
        print(f"   New shape: {df.shape} ‚Üí {df_clean.shape}")
    else:
        df_clean = df.copy()
        print(f"   ‚ö†Ô∏è  Keeping rows with missing dates (too many to remove)")
    
    return df_clean

cpi_final = handle_date_issues(cpi_final, "CPI")
commod_final = handle_date_issues(commod_final, "Commodity Prices")
```

```python
# Step 19: Final Dataset Summary
print("\nüìä FINAL DATASET SUMMARY")
print("="*50)

def final_summary(df, df_name):
    print(f"\nüéØ {df_name.upper()} - READY FOR ANALYSIS:")
    print(f"   Final shape: {df.shape}")
    print(f"   Date range: {df['date'].min()} to {df['date'].max()}")
    print(f"   Time span: {(df['date'].max() - df['date'].min()).days / 365.25:.1f} years")
    
    # Key columns available
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    print(f"   Key numeric columns: {[col for col in numeric_cols if not col in ['year', 'month', 'quarter']]}")
    
    # Remaining missing values
    remaining_missing = df.isna().sum().sum()
    missing_pct = (remaining_missing / (len(df) * len(df.columns))) * 100
    print(f"   Overall data completeness: {100-missing_pct:.1f}%")

final_summary(cpi_final, "CPI")
final_summary(commod_final, "Commodity Prices")
```
